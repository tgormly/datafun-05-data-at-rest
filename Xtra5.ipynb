## Renaming Columns



Right now, all of the column names are abbreviations, but they aren't descriptive.  We can use table.columns with a list of strings to rename all columns in the DataFrame to descriptive names.


```python
table.columns = [
    'Rank',
    'Player',
    'Age',
    'Position',
    'Games Played',
    'Goals',
    'Assists',
    'Points',
    '+/-',
    'Penalty Minutes',
    'EV','PP','SH','GW','EV','PP','SH','S','S%','TOI','ATOI','OPS','DPS','PS','BLK','HIT','FOW','FOL','FO%'
]
```

This notebook will not reference columns after "Penalty Minutes", so their column names are unchaged.

Let's see the updated table:

```python
table
```

This DataFrame contains 29 rows, one for each player.  The columns we are concerned with are:
<ul>
    <li>Rank - The player's overall rank on the team, as determined by a www.hockey-reference.com algorithm</li>
    <li>Player - The player's full name</li>
    <li>Age - The player's age</li>
    <li>Position - The position the player plays</li>
    <li>Games Played - How many games the player has participated in</li>
    <li>Goals - The number of goals the player has scored</li>
    <li>Assists - The number of goals a player has assisted in scoring</li>
    <li>Points - Total of goals and assists</li>
    <li>+/- - The difference between goals the player's team has scored while they are actively on the ice by goals the opposing team has scored while the player is actively on the ice</li>
    <li>Penalty Minutes - The number of penalty minutes the player has served</li>
</ul>

### Examining The Data

Now that our data is in order, we can perform some statistical analysis.

Let's run the describe method on a few statistics to see what the data tells us.

```python
table.Age.describe()
```

When we run the describe method on the Age column, we can see that: the mean age is 27.75; the oldest player is 37 years old, and the youngest player is only 20 years old; there is a standard deviation of 3.89 between the age of the players.

Because there is one row for each player, the count will always equal 29, the number of players on the team.

Let's check another statistic.

```python
table['Games Played'].describe()
```

When reviewing the number of Games Played, we see different data.  We might notice right away that there is a much greater standard deviation than in the age of the players at 15.99.  The number of games played by a player is a more disperse set of data than the ages of the players.  At most, a player has played in 52 games, while one player has played in only one game.  This player with one game is almost certainly an outlier - the lower quartile for this data set is 27.  This 27 is far away from 1, but quite near the mean of the data set.

```python
table.Goals.describe()
```

When reviewing the data in the Goals column, we can make a few observations.  The mean goals scored by the player on the St. Louis Blues is 5.48.  The minimum goals scored is zero.  Furthermore, the 25% quartile is 1, and the 50% quartile is only 3.  You can tell many of our players have struggled with offense this year.  There has been some success, however, as the 75% quartile breaks double digits with 10.  Our highest performer has scored 23 goals, well beyond the numbers supplied by his teammates.

## Displaying a Histogram

We can review a visual representation of some of this data in a histogram.  Let's take a look at the goals scored by our players first.  The Y axis represents the number of players who have scored this number of goals, and the X axis represents the number of goals scored.

```python
histogram = table.Goals.hist()
```

Goodness gracious, what a dismal picture.  By far, the number of goals scored with the greatest by the most players is zero or one.  Ideally, you would see many more players with 10 or more goals scored at this point in the season.

Does the picture improve when we look at total points?

```python
histogram = table.Points.hist()
```

Unfortunately, the two graphs look quite similar.
